```python
# Replace the filename with your uploaded file name
df = pd.read_csv('WORLD POPULATION SAMPLE.csv.csv', skiprows=4, delimiter=',')

# Preview the data
df.head()
```

```python
df_2020 = df.iloc[:,[0,63]].dropna()
df_2020.columns = ['Country Name', 'Popoulation']

non_countries = ['World', 'High income', 'Low income','European Union']
df_2020 = df_2020[~df_2020['Country Name'].isin(non_countries)]
df_2020.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,6))
plt.hist(df_2020['Popoulation'], bins = 30, color= 'skyblue', edgecolor = 'black')
plt.title('Population Distribution in 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```
